## Python List
List have specific ways to index it. In a list, elements inside can be add, remove, change, move, or copy through specific commands.

Take note that Python uses 0-based indexing. Meaning the first element is indexed as 0.

````
x = [1,2,3,6,5,4,7,8,9]

print(f'The first element is {x[0]} and the last element is {x[-1]}')
print(x[4])
print(x[4:7])                       # Note that the index 4 is inclusive and index 7 is exclusive
print(x[:7])                        # Include first element up to the 7th element
print(f'Elements in x = {len(x)}')  # len() returns the amount of elements in the list
````

There are also commands that can add, remove, or combine elements from the list.

````
x1 = [0,1,2,3,4]
x2 = [5,6,7,8,9]

# combine x1 and x2 list and assign to x3
x3 = x1 + x2
print(f'x3 = {x3}')

# Add element into the list
x1.append('apple')          # add 'apple' into x1 list
print(f'x1 = {x1}')
x1.insert(2,'fruit')        # add 'number' to 2nd index of the list
print(f'x1 = {x1}')

# change the 7th element to banana
x1[6] = 'banana'
print(f'x1 = {x1}')

# remove element from the list
last_ele = x1.pop()         # pop() remove the last element and return the element
print(f'Remove: {last_ele}, from: {x1}')
del x1[2]                   # remove the 3rd element
print(f'x1 = {x1}')
````

````
# 2D-array
x = [[1,2,3],[6,5,4],[7,8,9]]
print(x[0])
print(x[0][0])
````

## Loops (For and While)
### For Loops
A `for` loop is usually used for iterating over a list of data or to run a set of code for specified number of times.


````
x = [1,2,3,6,5,4,7,8,9]
for i in x:
  print(i)
````

`for` loop can be used with any sequence that is either a list, a tuple, a dictionary, a set, or a string.

````
animals = ['dog','cat','hamster','monkey']
for x in animals:
  print(x)
````

````
for i in range(10):
  print(i)
````

````
# A nice way to iterate over the items is by using enumerate
for i,j in enumerate(animals):
  print(f"Index: {i}, Value: {j}")
````

### While Loops
A `while` loop is used for looping a set of code as long as the specified condition(s) is still true.

````
i = 1
while i < 6:
  print(i)
  i += 1
````

````
i = 1
while True:
  print(i)
  i += 1
  if i == 6:
    break
````

## lambdas and map

### Lambdas
A lambda function is a small anonymous function. It can take any number of arguments but can only have one expression.


````
add = lambda x, y : x + y
a = 12
print(add(5,6))
````

A lambda function is really useful for creating an anonymous function inside another funciton.

````
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)
# f(x) = 2x
print(f'mydoubler returns: {mydoubler(8)}')

mytrippler = myfunc(3)
print(f'mytrippler returns: {mytrippler(8)}')
````

### Map
Map allows you to run a piece of function to all of the arguments passed. Map function (mydoubler) applies to each element in the array.


````
def myfunc(n):
  return n*n

result = map(myfunc, [1,2,3,4])
print(list(result))
````

````
dict_a = [{'name': 'student_a', 'points': 100}, {'name': 'student_b', 'points': 98}]
c = {'a':1,'b':2}
print(list(map(lambda x : x['name'], dict_a)))
````

## An introduction to Object Oriented Programming in Python
Object Oriented Programming is a paradigm in which we store data and methods that operate on that data. Imagine a class as a blueprint that contains all the variables and methods to build that particular object.
OOPs is all about creating objects that have data and functions associated with it.<br>
For example: 
A parrot can be an object,as it has the following properties:

- name, age, color as attributes
- singing, dancing as behavior
The concept of OOP in Python focuses on creating reusable code. This concept is also known as DRY (Don't Repeat Yourself).

<img src = "https://nbisweden.github.io/RaukR-2019/oop_Marcin/oop_presentation/oop_assets/oop_concept.png">

````
# class is defined using the class keyword
# Constructors are defined using __init__ method
# self is the same as this in Java (that is the current instance)
class Person:

  # Constructor
  def __init__(self, name, age):
    self.name = name
    # this.name = name;
    self.age = age
    # this.age = age;
  
  # Get age after (after years)
  def after(self, after):
    return self.age + after
  
  # Display the results
  def display(self):
    return f'{self.name}:{self.age}'
  
  # ToString() method like in Java
  def __str__(self):
    return self.name + str(self.age)

p = Person("a", 12)

p2 = Person("a", 1234)


print(p.after(100))
print(p.display())
````

## List comprehensions
According to Wikipedia, A list comprehension is a alternate way available in some programming languages for creating a list based on existing lists.

To summarise, list comprehensions are a syntactic sugar (a short hand way of doing things) over normally writing multiple lines of code. This feature is largely inspired by functional programming (which you would learn more about in College CS courses)

````
# Normal way (7 lines)
l = [1, 2, 3, 4, 5]
even = []
for i in l:
  if i%2 == 0:
    even.append(i)
print("Even nums: ", even)


# List comprehensions (2 lines)
evens = [i for i in l if i % 2 == 0]
print("Even nums: ", evens)
````

**As you can see, using list comprehensions we were able to achieve the same result with just 2 lines of code instead of using 7 lines of code.**
